```python
import pandas as pd
import numpy as np

from datetime import datetime

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
sns.set(style='darkgrid')

mpl.rcParams['figure.figsize'] = (16,9)
pd.set_option('display.max_rows', 500)
```

```python
dataPath_Processed = ("../data/processed/COVID_Flat_Table_small.csv")
```

```python
df_analyse = pd.read_csv(dataPath_Processed, sep=',')
```

```python
df_analyse.sort_values('date',ascending=True).head()
```

```python
country = 'Germany'

N0 = 1000000
beta = 0.4
gamma = 0.1

I0 = df_analyse.Germany[100]
S0 =  N0-I0
R0 = 0
```

```python
def SIR_model(SIR,beta, gamma):
    
    S,I,R = SIR
    dS_dt = -beta*S*I/N0
    dI_dt = beta*S*I/N0-gamma*I
    dR_dt = gamma*I
    
    return([dS_dt,dI_dt,dR_dt])
```

```python
SIR = np.array([S0,I0,R0])

propagation_rate = pd.DataFrame(columns={'Susceptible':S0,
                                        'Infected':I0,
                                        'Recovered':R0
                                        })

for each_t in np.arange(50):
    
    new_delta_vec = SIR_model(SIR,beta,gamma)
    SIR = SIR + new_delta_vec
    propagation_rate = propagation_rate.append({'Susceptible':SIR[0],
                                        'Infected':SIR[1],
                                        'Recovered':SIR[2]
                                        }, ignore_index = True)
```

```python
fig, ax1 = plt.subplots(1,1)

ax1.plot(propagation_rate.index, propagation_rate.Infected, label='Infected', color='k')
ax1.plot(propagation_rate.index, propagation_rate.Recovered, label='Recovered')
ax1.plot(propagation_rate.index, propagation_rate.Susceptible, label='Susceptible')

ax1.set_ylim(10, 1000000)
ax1.set_yscale('linear')
ax1.set_title('SIR Simulations', size = 16)
ax1.legend(loc='best',
          prop={'size': 16})
```

# Fitting the parameters of SIR model

```python
from scipy import optimize
from scipy import integrate

ydata = np.array(df_analyse.India[100:])
t=np.arange(len(ydata))
```

```python
I0 = ydata[0]
S0 = 1000000
R0 = 0
beta
```

```python
def SIR_model(SIR,t,beta, gamma):
    
    S,I,R = SIR
    dS_dt = -beta*S*I/N0
    dI_dt = beta*S*I/N0-gamma*I
    dR_dt = gamma*I
    
    return dS_dt,dI_dt,dR_dt
```

```python
def fit_odeint(x, beta, gamma):
    return integrate.odeint(SIR_model, (S0,I0,R0), t, args=(beta,gamma))[:,1]
```

```python
popt = [0.4,0.1]
fit_odeint(t, *popt)
```

```python
popt,pcov = optimize.curve_fit(fit_odeint, t, ydata, maxfev=100000)
perr = np.sqrt(np.diag(pcov))

print('Standard deviation errors :', str(perr), ' start infect:', ydata[0])
print('Optimal parameters: beta = ', popt[0], ' and gamma = ', popt[1])
```

```python
fitted = fit_odeint(t, *popt)
```

```python
plt.semilogy(t,ydata, 'o')
plt.semilogy(t, fitted)
plt.title("Fit of SIR model for cases in %s" %country)
plt.ylabel('Infected Population')
plt.xlabel('Number of Days')
plt.show()
print('Optimal Parametes: beta = ', popt[0], ' and gamma = ', popt[1])
```
